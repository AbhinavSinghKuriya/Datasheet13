```python
# Boxplot for outlier detection

numerical_cols = ['Pclass', 'Age', 'SibSp',
                  'Parch', 'Fare']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```
